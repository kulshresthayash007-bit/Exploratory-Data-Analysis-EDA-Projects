```python
bool_cols = ['paywall_paid', 'desktop_enter', 'is_big_city']
for col in bool_cols:
    df[col] = df[col].astype(int) 
```

```python
date_cols = ['first_trial_appointment_date', 'first_payment_date',
             'call_date', 'first_visit_date'
            ]
for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

```python
fill_cols = ['os', 'tutor', 'job', 'school_name', 'region']
df[fill_cols] = df[fill_cols].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix (ensuring only numeric data is used)
df_corr = df.corr(numeric_only=True) 

# Plot the heatmap and annotations
fig, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(
    df_corr, 
    xticklabels=df_corr.columns, 
    yticklabels=df_corr.columns, 
    annot=True, 
    fmt=".2f",    
    cmap="Spectral", 
    ax=ax
)
plt.show()
```
